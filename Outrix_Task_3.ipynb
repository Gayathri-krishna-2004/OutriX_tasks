```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load dataset
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn-2.csv")
```

```python

# Clean and preprocess
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(subset=['TotalCharges'], inplace=True)
df['SeniorCitizen'] = df['SeniorCitizen'].replace({1: 'Yes', 0: 'No'})
```

```python

# Churn counts
print("Churn Counts:\n", df['Churn'].value_counts())
print("\nChurn Rate (%):\n", df['Churn'].value_counts(normalize=True) * 100)
```

```python

# Plot: Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title("Customer Churn Count")
plt.show()
```

```python

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['tenure', 'MonthlyCharges', 'TotalCharges']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Tenure vs Churn
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title("Tenure Distribution by Churn")
plt.xlabel("Tenure (Months)")
plt.show()
```

```python
# Monthly Charges vs Churn
plt.figure(figsize=(10,6))
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True)
plt.title("Monthly Charges Distribution by Churn")
plt.xlabel("Monthly Charges")
plt.show()
```

```python
# Contract Type vs Churn
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set1')
plt.title("Churn by Contract Type")
plt.show()
```

```python
# Internet Service vs Churn
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set3')
plt.title("Churn by Internet Service Type")
plt.show()
```
